## 常规文件读写

### CSV

```python
import pandas as pd
```

```python
# 读
data = pd.read_csv('data/movie.csv')
data.head()
```

```python
# 写
data[:10].to_csv('data/gen_data.csv', columns=['color', 'director_name'])
```

```python
pd.read_csv('data/gen_data.csv').head()
```

### HDF5

```python
# 写 必须指定key
data[:10].to_hdf('data/gen_data_2.h5', key='hdfkey')
```

```python
pd.read_hdf('data/gen_data_2.h5', key='hdfkey').head()
```
