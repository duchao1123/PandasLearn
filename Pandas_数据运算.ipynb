## 算数运算
- add
- sub

```python
arr = np.array([
    [1, 2, 3], 
    [4, 5, 6]])
df = pd.DataFrame(arr)
df
```

```python
df.add(3)
```

```python
df.sub(1)
```

### 逻辑运算
- data[1] > 2 返回逻辑结果
- data[data[1] > 2] 根据逻辑结果筛数据
- 多逻辑判断 (data[1] > 3) & (data[1] < 6)
- query('open<24 & open>23')
- isin([1, 2, 3])

```python
df[df[1] > 2]
```

```python
df
```

```python
# 合并数据
s = pd.DataFrame([[7, 8, 9]])
pd.concat([df, s], ignore_index=True)
```

```python
df[(df[1] > 3) & (df[1] < 6)]
```
