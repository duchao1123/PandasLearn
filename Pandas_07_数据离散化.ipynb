```python
import pandas as pd
import numpy as np
```

## 数据离散化
- cut自定义区间分组
- qcut自动区间分组
- get_dummies对数据one-hot编码(整行只有一个1，其余都是0)

```python
data = pd.read_csv('data/nobel_prizes.csv')
data.head()
```

```python
data.describe()
```

```python
year = data['year']
year
np.any(year.isnull())
```

```python
# 自动分组
qcut = pd.qcut(year, 10)
qcut
```

```python
# 统计各分组数量
qcut.value_counts()
```

```python
# 自定义分组
cut = pd.cut(year, [1900, 1936, 2010])
cut.value_counts()
```

```python
dummies = pd.get_dummies(data, prefix='rise')
dummies
```
