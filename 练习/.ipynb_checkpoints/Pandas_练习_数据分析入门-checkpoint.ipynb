```python
import pandas as pd

movie = pd.read_csv('../data/movie.csv')
movie.head()
```

## 需求找出小成本、高口碑电影

```python
# 高口碑100 -> 低成本5条
movie.nlargest(100, 'imdb_score').nsmallest(5, 'budget')
```

## 每年imdb评分最高的电影

```python
# 取出使用值
movies = movie[['movie_title', 'title_year', 'imdb_score']]
```

```python
# 对title_year， imdb_score排序，倒序
sorted_movies = movies.sort_values(['title_year', 'imdb_score'], ascending=False)
```

```python
# 针对title_year去重，获取第一个值
sorted_movies.drop_duplicates(subset=['title_year'], keep='first')
```
