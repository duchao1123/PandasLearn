### 二维数据 DataFrame

```python
# 1、创建
```

```python
# 由二维list创建
a = pd.DataFrame([
    [1, 1, 1],
    [2, 2, 2]
])
print(type(a))
a
```

```python
# 由ndarray创建
import numpy as np
arr = np.array([
    [1, 1, 1],
    [2, 2, 2]
])
b = pd.DataFrame(arr)
print(type(b))
b
```

```python
# 指定列索引
arr_1 = np.array([
    ['x', 18, '23001'],
    ['y', 16, '23002']
])
b = pd.DataFrame(arr, columns=['c_1', 'c_2', 'c_3'])
b
```

```python
# 3、常用显示操作
```

```python
arr_2 = np.random.randint(0, 100, size=(10, 10))
df = pd.DataFrame(arr_2, columns=[f'col_{i}' for i in range(1, 11)])
```

```python
# 显示前5行, 默认5
df.head()
```

```python
# 显示前8行
df.head(8)
```

```python
# 显示后n行
df.tail(2)
```

```python
# 3、索引操作
```

```python
# 修改索引
df.index = [f'row_{i}' for i in range(1, df.index.size + 1)]
df
```

```python
# 重置默认索引，或删除当前索引
new_df = df.reset_index(drop=True)
new_df
```

```python
# 原数据不受影响
df
```

```python
# 指定索引
new_index_df = df.set_index('col_1', drop=True)
new_index_df
```
