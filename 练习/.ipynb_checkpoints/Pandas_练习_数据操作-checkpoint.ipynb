```python
import pandas as pd
import numpy as np
```

```python
# 模拟一个100 * 100 正态分布的数据
data1 = pd.DataFrame(np.random.normal(1.2, 1, (100, 100)))
data2 = pd.DataFrame(np.ones((100, 100)))
```

```python
# 加
data1.add(data2).head()
```

```python
# 乘
data1.multiply(data2).head()
```

```python
# 使用逻辑运算过滤数据
# 将所有小于0的数据设置为0，无效数据
data1[data1 < 0] = 0
data1.head()
```

```python
# isin, query 过滤
data2.isin([1, 2, 3]).head()
```

```python
# 综合统计
data1.describe()
```

```python
data1.cumsum(axis=1).head()
```

```python
# 自定义运算 apply(func, axis=0)

def myfun(x):
    return x ** 2 + 1

data1.apply(myfun, axis=1)
```
