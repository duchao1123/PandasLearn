```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movie = pd.read_csv('../data/movie.csv')
movie.head()
```

### 需求：分析统计电影类型信息(genres)

#### 需求分析：
1. 获取所有行的genres数据
2. 遍历提取所有genres数据
3. 生成统计genres的新表数据
4. 如果属于对应类型赋值为1，否则赋值为0
5. 统计数据
6. 图像化数据

```python
genres = movie['genres']
genres.head()
```

```python
# 1、获取每行数据的genres
rows_genres = [item.split('|') for item in genres]

# 2、按照分割符拆分每行的genres
one_genres = [one for row in rows_genres for one in row]

# 3、去重
uni_genres = pd.unique(one_genres)
uni_genres
```

```python
genres_table = np.zeros((genres.shape[0], uni_genres.shape[0]))
genres_df = pd.DataFrame(genres_table, columns=uni_genres)
genres_df.head()
```

```python
index = 0
for row in rows_genres:
    for c in row:
        genres_df[c][index] = 1
    index += 1
```

```python
genres_df.head()
```

```python
genres_df.describe()
```

```python
# 求和并排序数据
sort_genres = genres_df.sum().sort_values(ascending=False)
sort_genres
```

```python
# 图像化数据
# 排序型数据，使用条形图
sort_genres.plot(kind='bar', figsize=(20, 8), fontsize=16, colormap='cool')
```
